```python
data = data.dropna(subset=['Thermal Conductivity'])
```

### Sorting the table according to increasing order of thermal conductivities

```python
data = data.sort_values('Thermal Conductivity')
data
```

```python
def binning(col):
    min_ = min(data[col])
    max_ = max(data[col])
    bin_range = (max_ - min_)//7
    bins = []
    for bin_ in range(min_, max_, int(bin_range)):
        bins.append(bin_)
    return bins
```

```python
bins = binning('Thermal Conductivity')
bins
```
